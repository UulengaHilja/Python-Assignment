```python
#Two operands using different operators in python
a = 10
b = 5

# Addition
result = a + b
print(f"Addition of {a} and {b} is {result}")   # Output: Addition of 10 and 5 is 15

# Subtraction
result = a - b
print(f"Subtraction of {a} and {b} is {result}")   # Output: Subtraction of 10 and 5 is 5

# Multiplication
result = a * b
print(f"Multiplication of {a} and {b} is {result}")   # Output: Multiplication of 10 and 5 is 50

# Division
result = a / b
print(f"Division of {a} and {b} is {result}")   # Output: Division of 10 and 5 is 2.0

# Modulus
result = a % b
print(f"Modulus of {a} and {b} is {result}")   # Output: Modulus of 10 and 5 is 0

# Exponentiation
result = a ** b
print(f"{a} raised to the power of {b} is {result}")   # Output: 10 raised to the power of 5 is 100000

```

```python
# Creating objects of different data types with dunder/magic methods and other attributes

# Integer
a = 5
print("Integer:")
print(dir(a))   # Get list of all attributes and methods for the integer object

print(a.__add__(5))   # Output: 10, same as a+5
print(a.__mul__(2))   # Output: 10, same as a*2
print(type(a))        # Output: <class 'int'>
print(isinstance(a, int))   # Output: True

# Float
b = 3.14
print("\nFloat:")
print(dir(b))   # Get list of all attributes and methods for the float object

print(b.__add__(1.86))   # Output: 5.0, same as b+1.86
print(b.__truediv__(2))  # Output: 1.57, same as b/2
print(type(b))           # Output: <class 'float'>
print(isinstance(b, float))   # Output: True

# String
c = "Hello, World!"
print("\nString:")
print(dir(c))   # Get list of all attributes and methods for the string object

print(c.__len__())        # Output: 13, same as len(c)
print(c.__contains__("o"))   # Output: True, same as "o" in c
print(type(c))           # Output: <class 'str'>
print(isinstance(c, str))   # Output: True

# List
d = [1, 2, 3, 4, 5]
print("\nList:")
print(dir(d))   # Get list of all attributes and methods for the list object

d.__setitem__(2, 6)   # Same as d[2] = 6
print(d)   # Output: [1, 2, 6, 4, 5]
d.__delitem__(-1)     # Same as del d[-1]
print(d)   # Output: [1, 2, 6, 4]
print(type(d))        # Output: <class 'list'>
print(isinstance(d, list))   # Output: True

# Tuple
e = (1, 2, 3, 4, 5)
print("\nTuple:")
print(dir(e))   # Get list of all attributes and methods for the tuple object

print(e.__getitem__(2))   # Output: 3, same as e[2]
print(type(e))            # Output: <class 'tuple'>
print(isinstance(e, tuple))   # Output: True

# Dictionary
f = {"name": "Hilja", "age": 30, "city": ""}
print("\nDictionary:")
print(dir(f))   # Get list of all attributes and methods for the dictionary object

print(f.__getitem__("name"))   # Output: "Hilja", same as f["name"]
f.__setitem__("age", 31)       # Same as f["age"] = 31
print(f)   # Output: {"name": "Hilja", "age": 31, "city": "Berlin"}
print(type(f))        # Output: <class 'dict'>
print(isinstance(f, dict))   # Output: True

# Set
g = {1, 2, 3, 4, 5}
print("\nSet:")
print(dir(g))   Berlin# Get list of all attributes and methods for the set object

g.__add__({6, 7, 8})     
```

```python
#Index in pythons
# Create a list of countries
countries = ["Finland", "Sweden", "Denmark", "Netherlands", "Germany", "Norway", "Iceland"]

# Find the index of "Sweden"
print(countries.index("Sweden"))   # Output: 1

# Get all the items except the first
print(countries[1:])   # Output: ['Sweden', 'Denmark', 'Netherlands', 'Germany', 'Norway', 'Iceland']

# Get all the items except the last
print(countries[:-1])   # Output: ['Finland', 'Sweden', 'Denmark', 'Netherlands', 'Germany', 'Norway']

# Find the last item using negative indexing
print(countries[-1])   # Output: 'Iceland'

# Find the last three items using negative indexing
print(countries[-3:])   # Output: ['Germany', 'Norway', 'Iceland']

# Grab all the items with odd indices
print(countries[1::2])   # Output: ['Sweden', 'Netherlands', 'Norway']

# Find the first and last item for the list
print(countries[0])   # Output: 'Finland'
print(countries[-1])   # Output: 'Iceland'

# Get ['Netherlands', 'Germany', 'Norway'] using negative indexing
print(countries[-4:-1])   # Output: ['Netherlands', 'Germany', 'Norway']

# Reverse the list
print(countries[::-1])   # Output: ['Iceland', 'Norway', 'Germany', 'Netherlands', 'Denmark', 'Sweden', 'Finland']

# What is the output of countries[::-2]?
print(countries[::-2])   # Output: ['Iceland', 'Germany', 'Denmark', 'Finland']

# What is the output of countries[1:100:2]
print(countries[1:100:2])  # Output: ['Sweden', 'Netherlands', 'Norway']

# What is the output of countries[len(countries)-1:0:2]?
print(countries[len(countries)-1:0:2])  # Output: []

# Get the output ['Germany', 'Netherlands', 'Denmark']
print(countries[-4:-1])  # Output: ['Netherlands', 'Germany', 'Norway']

# Use enumerate method to get index and values in the list
for i, country in enumerate(countries):
    print(i, country)
    
    #Output
0 Finland
1 Sweden
2 Denmark
3 Netherlands
4 Germany
5 Norway
6 Iceland
```

```python
# Given the two lists of same length, create a dictionary using a list as keys and another list as values without using loop
keys = ["a", "b", "c", "d"]
values = [1, 2, 3, 4]
dictionary = dict(zip(keys, values))
print(dictionary)  # Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4}

# Create two dictionaries, merge them into one.
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
merged_dict = dict1.copy()
merged_dict.update(dict2)
print(merged_dict)  # Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4}

# Given the list of students, assign them the course.
students = ["Duy", "Laxmi", "Antonio", "Maria"]
courses = ["Python", "Database", "ML"]
students_profile = {student: courses for student in students}
print(students_profile)  # Output: {'Duy': ['Python', 'Database', 'ML'], 'Laxmi': ['Python', 'Database', 'ML'], 'Antonio': ['Python', 'Database', 'ML'], 'Maria': ['Python', 'Database', 'ML']}

# Modify Antonio's course to ["JavaScript", "Database", "NodeJS"]
students_profile["Antonio"] = ["JavaScript", "Database", "NodeJS"]
print(students_profile)  # Output: {'Duy': ['Python', 'Database', 'ML'], 'Laxmi': ['Python', 'Database', 'ML'], 'Antonio': ['JavaScript', 'Database', 'NodeJS'], 'Maria': ['Python', 'Database', 'ML']}

# Add "ReactJs" to Antonio's course
students_profile["Antonio"].append("ReactJs")
print(students_profile)  # Output: {'Duy': ['Python', 'Database', 'ML'], 'Laxmi': ['Python', 'Database', 'ML'], 'Antonio': ['JavaScript', 'Database', 'NodeJS', 'ReactJs'], 'Maria': ['Python', 'Database', 'ML']}

# Replace Antonio's "ReactJs" to "Vuejs"
students_profile["Antonio"][students_profile["Antonio"].index("ReactJs")] = "Vuejs"
print(students_profile)  # Output: {'Duy': ['Python', 'Database', 'ML'], 'Laxmi': ['Python', 'Database', 'ML'], 'Antonio': ['JavaScript', 'Database', 'NodeJS', 'Vuejs'], 'Maria': ['Python', 'Database', 'ML']}

# Remove an item from dictionary
del students_profile["Maria"]
print(students_profile)  # Output: {'Duy': ['Python', 'Database', 'ML'], 'Laxmi': ['Python', 'Database', 'ML'], 'Antonio': ['JavaScript', 'Database', 'NodeJS', 'Vuejs']}
```
